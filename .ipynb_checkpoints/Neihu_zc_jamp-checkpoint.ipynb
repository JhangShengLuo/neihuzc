```python

#讀兩個檔案把他們合併成內湖跟中正的index table

nhzcindex=pd.read_excel('neihuZC.xlsx',encoding='utf-8');
nhzcindexc=nhzcindex.copy()

pattern= {'direct':{u'.?東$':'E',u'.?西$':'W',u'.?南$':'S',u'.?北$':'N',u'.向放置':'U',u'未指定':'U'},\
         'newdirect':{u'.?東.?':'E',u'.?西.?':'W',u'.?南.?':'S',u'.?北.?':'N',u'.向放置':'U',u'未指定':'U'}}

nhzcindexc=nhzcindexc.replace(pattern,regex=True);


nhzclist = ['NeiHuVD.csv','ZhongChengVD.csv']


frames= []
for table in nhzclist:
    temp = pd.read_csv(table)
    temp.columns=['tablename']
    frames.append(temp)

nhzctable = pd.concat(frames,ignore_index=True)    
nhzctablec = nhzctable.copy()


nhzctablec['VD'] = nhzctablec.ix[:,0].str[1:]
nhzctablec['direct'] = nhzctablec.ix[:,0].str[:1];

nhzcresult=pd.merge(nhzctablec, nhzcindexc, on=['VD','direct'])
nhzcresult.sortby = nhzcresult.index
nhzcresult.to_csv('nhzcindex.csv',index=False)
```

```python
nhzcresultc=nhzcresult.copy()
```

```python
nhzc_table_list=nhzcresult.ix[:,0].tolist()
```

```python
import time
tStart = time.time()

db_connection = sql.connect(host='localhost', database='taipeiall', user='root', password='apple')
frames = []
for ele in nhzc_table_list:
    frames.append(pd.read_sql('SELECT * FROM {r}'.format(r=ele), con=db_connection))

result = pd.concat(frames, ignore_index=True)

tEnd = time.time()
print "It cost %f sec" % (tEnd - tStart)
```

```python
# result
def decode_to_utf8(df):
    df.EQIPnumber=df.EQIPnumber.str.decode("utf-8")
    df.location=df.location.str.decode("utf-8")
    df.direct=df.direct.str.decode("utf-8")
    df.YMD=df.YMD.str.decode("utf-8")
    df.hour=df.hour.str.decode("utf-8")
    return df

df_all=decode_to_utf8(result) 
```

```python
df_all=df_all.drop(['scooterflow','truckflow','addTotal','carflow','avgCarSpace','laneNumber'],axis=1)
```

```python
df_all
pattern= {'direct':{u'.?東$':'E',u'.?西$':'W',u'.?南$':'S',u'.?北$':'N',u'.向放置':'U',u'未指定':'U'}}

df_all=df_all.replace(pattern,regex=True);
```

```python
df_all['YMD']=pd.to_datetime(df_all['YMD']);

df_all['weekday'] = df_all['YMD'].dt.dayofweek;
```

```python
hour_to_num={'00:00:00':0,'01:00:00':1,'02:00:00':2,'03:00:00':3,'04:00:00':4,'05:00:00':5,'06:00:00':6,'07:00:00':7\
            ,'08:00:00':8,'09:00:00':9,'10:00:00':10,'11:00:00':11,'12:00:00':12,'13:00:00':13,'14:00:00':14\
            ,'15:00:00':15,'16:00:00':16,'17:00:00':17,'18:00:00':18,'19:00:00':19,'20:00:00':20\
             ,'21:00:00':21,'22:00:00':22,'23:00:00':23}

pattern= {'hour':hour_to_num}

df_all=df_all.replace(pattern);


```

```python
df_all=df_all.drop(['laneNumber'],axis=1)
df_all.columns = ['VD','location','direct','YMD','time','eq','speed','accu','weekday']
```

```python
nhzcsortby=nhzcresultc.drop(['tablename','road','position','city','district','district','newdirect'],axis=1)
```

```python

tempmerge=pd.merge(nhzcsortby, df_all, on=['VD','direct'])
```

```python
predict_template=pd.read_excel('predict_template.xlsx',encoding='utf-8');
template = predict_template.copy()
```

```python
sortbylist=nhzcresultc.sortby.tolist()
tempframe = []

for i in sortbylist:
    print i
    template = predict_template.copy()
    wwp=tempmerge.where(tempmerge.sortby==i).dropna().copy()

    sp45=wwp.speed.quantile(0.4);sp45
    wwpcjam=wwp.where((wwp.speed<sp45)&(wwp.accu>5)).dropna().copy()

    template.ix[:,'sortby']=i;
    for dd,tt in template[['weekday','time']].values.tolist():
        alldaynum=wwp[(wwp.weekday==dd)&(wwp.time==tt)].count()[0]
        jamdaynum=wwpcjam[(wwpcjam.weekday==dd)&(wwpcjam.time==tt)].count()[0]
        if alldaynum==0:
            template.ix[(template.time==tt)&(template.weekday==dd),'jamp']=0
    #         print '\n'
    #         print "星期",dd+1,"_",tt,'時'
    #         print '總共',alldaynum,'週','塞車週數',jamdaynum
        else:
            template.ix[(template.time==tt)&(template.weekday==dd),'jamp']=round((float(jamdaynum)/alldaynum),4);template


    tempframe.append(pd.merge(template,nhzcresultc,on=['sortby'],copy=False))


```

```python
nhzctest=pd.concat(tempframe, ignore_index=True);
nhzctest.to_csv('NeihuZhongcheng.csv',index=False)
```
